**Preparation 1:**  How profitable are the product categories defined by ROI?

```python
# Preparation 1

# Define product categories column
# This slices off the 'PROD_' prefix from each Product_ID (which is 5 characters long) 
# and stores the remainder in a new column called Product_Category. 
# It’s a clean way to isolate the unique identifier or suffix from each product code.

mapp_df['Product_Category'] = mapp_df['Product_ID'].str[5:]
mapp_df[['Product_ID', 'Product_Category']].head(10)

# Define bins and labels
# This defines thresholds for grouping the data

bins = [0, 3.99, 4.99, 5.0]
labels = ['Low ROI (0–3.99)', 'Mid ROI (4–4.99)', 'High ROI (5.0)']

# Create the tiered ROI column:
# This segments the continuous ROI values into labeled bins (tiers) using bins and labels defined above
# This also adds the result as a new column called ROI_Tier. 
# This makes it easier to compare performance across the tiers later.

mapp_df['ROI_Tier'] = pd.cut(mapp_df['ROI'], bins=bins, labels=labels, include_lowest=True)

```

**Preparation 2:** Are customers more likely to be in a higher subscription tier given a higher subscription length?

```python
# Preparation 2

bins = [0, 3, 6, 12, 24, mapp_df['Subscription_Length'].max()]
labels = ['0–3 mo', '4–6 mo', '7–12 mo', '13–24 mo', '25+ mo']
```

**Preparation 3:** How does advertising budget impact customer engagement across campaigns? 

```python
# Preparation 3

# [Budget, Clicks, and Conversions were confirmed to be numeric]

# Prevent dividing by zero:
mapp_df = mapp_df[mapp_df['Clicks'] > 0].dropna(subset = ['Budget', 'Clicks', 'Conversions'])

# Adding conversion rate column:
mapp_df['Conversion_Rate'] = mapp_df['Conversions'] / mapp_df['Clicks']
```

### 4. Analyze the data

**Analysis 1:** How profitable are the product categories defined by ROI? 

**Answer:** The vast majority of the total revenue comes from investments that have a relatively low ROI. <br/>
The investments with a high ROI contribute very little to the total revenue shown. <br/>
This could imply a strategy where a large volume of low-ROI activities are pursued, or that high-ROI opportunities are rare or not scaled up to generate significant overall revenue. <br/>

```python
# Analysis 1: 

# Group by ROI tier
# This groups the dataset by ROI_Tier, and aggregates Total Revenue_Generated, Total Units_Sold:
roi_tier_summary = mapp_df.groupby('ROI_Tier', observed=True).agg({
    'Revenue_Generated': 'sum',
    'Units_Sold': 'sum',
    'Product_ID': 'nunique'
}).reset_index()

# Rename the column for clarity - it contains the count of unique product_ID's within the ROI_Tier:
roi_tier_summary.rename(columns={'Product_ID': 'Unique_Products'}, inplace = True)

# Verify result:
roi_tier_summary
```

**Analysis 2:** Are customers more likely to be in a higher subscription tier given a higher subscription length?

**Answer:** The "Basic" subscription tier generally has the highest percentage of customers across almost all subscription lengths, indicating it might be the most popular or accessible option.
The distribution of customers across the three tiers is relatively even, with each tier typically holding between 31% - 36% of customers in any subscription length category.
The "Standard" tier shows a slightly higher proportion of customers in the shortest subscription length (0-3 months), suggesting it's a popular starting point for new subscribers or for those trying out the service. 

As subscription length increases, the "Basic" tier tends to be selected.

```python
# Analysis 2: 

# Use cut to create bins and labels, then group by Length Bin and Subscription Tier:
mapp_df['Length_Bin'] = pd.cut(mapp_df['Subscription_Length'], bins = bins, labels = labels)
grouped = mapp_df.groupby(['Length_Bin', 'Subscription_Tier'], observed = False).size().reset_index(name='Count')

# Normalize by length bin:
total_per_bin = grouped.groupby('Length_Bin', observed = False)['Count'].transform('sum')
grouped['Percent'] = grouped['Count'] / total_per_bin

# Verify result:
grouped
```

**Analysis 5:** How does Revenue Generated change over Subscription Length for each Subscription Tier?<br/>

**Answer:** This data shows the dynamic nature of total revenue generated by subscription length for each tier. <br/>
There's no clear linear relationship, and the performance of each tier varies a lot depending on the subscription duration. <br/>
Businesses would need to analyze specific peaks and troughs for each tier to understand what might be driving these revenue changes at particular subscription lengths. <br/>

A dip might indicate where customers churn, while a peak might suggest a successful retention strategy or a popular initial subscription.<br/>

```python
# Analysis 5: 

# Group by 'Subscription_Length' and 'Subscription_Tier' and sum 'Revenue_Generated'
revenue_by_length_tier = mapp_df.groupby(['Subscription_Length', 'Subscription_Tier'])['Revenue_Generated'].sum().reset_index()

# Sort by Subscription_Length for a clearer line graph
revenue_by_length_tier = revenue_by_length_tier.sort_values(by='Subscription_Length')

print("\nAnalyzed Data (Revenue by Subscription Length and Tier):")
print(revenue_by_length_tier.head())
```

**Analysis 6:** What common keywords are most associated with high-performing products? <br/>

**Answer:** The affordability is the most common characteristic among high-performing products, followed closely by innovation and durability. <br/>
Style also plays a role but to a slightly lesser extent than the other three. <br/>

This data suggests that when aiming for high-performing products, aesthetics are not the primary focus!<br/>

```python
# Analysis 6: 

# Clean the 'Common_Keywords' column directly
# Fill NaN values with an empty string before applying string methods:
high_performing_df['Common_Keywords_Cleaned'] = high_performing_df['Common_Keywords'].fillna('').str.strip().str.lower()

# Count the frequency of each keyword using value_counts():
keyword_counts_hp = high_performing_df['Common_Keywords_Cleaned'].value_counts()

# Remove any empty string entry if it exists (from NaN fill or empty strings in original data)
if '' in keyword_counts_hp.index:
    keyword_counts_hp = keyword_counts_hp.drop('')

# Get the most common keywords (Series of counts)
# Will display top 10 or fewer if there aren't that many unique keywords:
num_keywords_to_show = min(10, len(keyword_counts_hp))
most_common_keywords_hp_series = keyword_counts_hp.head(num_keywords_to_show)

print(f"\nTop {num_keywords_to_show} Common Keywords in High-Performing Products (by Frequency):")
print(most_common_keywords_hp_series)
```

### 5. Visualize the data

**Visualization 1: Barplot** - Which are the most profitable product categories defined by ROI? 

```python
# Visualization 1: 

# Most products have an ROI of 5, but they generate far less revenue and sales volume than lower tiers
# This suggests that high ROI products may be niche or under-leveraged!

plt.figure(figsize=(10, 6))

barplot = sns.barplot(
    data = roi_tier_summary,
    x = 'ROI_Tier',
    y = 'Revenue_Generated',
    hue='ROI_Tier',
    palette='YlGnBu', # Making it pretty
    edgecolor = 'black',
    legend = False)

# Add value labels on top of bars
for container in barplot.containers:
    barplot.bar_label(container, fmt='${:,.0f}', padding=5, fontsize=11, weight='bold')

# Customize axes and titles
plt.title('Total Revenue by ROI Tier', fontsize = 16, fontweight = 'bold', pad = 15)
plt.xlabel('ROI Tier', fontsize = 12)
plt.ylabel('Revenue Generated', fontsize = 12)
plt.xticks(fontsize = 11)
plt.yticks(fontsize = 11)
plt.show()
```

**Visualization 2: Barplot -** Are customers more likely to be in a higher subscription tier given a higher subscription length?

```python
# Visualization 2: 

# Customers who commit long-term may prioritize affordability over premium features -- especially if the Basic tier meets their core needs
# Premium subscribers might trial short-term and possibly churn before upgrading/extending
# The Standard tier holds steady (possibly offering enough value to maintain interest at every length?)

plt.figure(figsize=(10,6))

sns.barplot(data=grouped, 
            x='Length_Bin', 
            y='Percent', 
            hue='Subscription_Tier', 
            palette='YlGnBu',  # Making it pretty
            edgecolor='black')

plt.title('Subscription Tier Distribution Across Subscription Lengths')
plt.xlabel('Subscription Length')
plt.ylabel('Percentage of Customers')

plt.legend(
    title='Subscription Tier', 
    bbox_to_anchor=(1.05, 1), 
    loc='upper left')

plt.tight_layout()
plt.show()
```

**Visualization 3: Scatterplot** - How does advertising budget impact customer engagement across campaigns? 

```python
# Visualization 3: 

# The biggest observation is the lack of any discernible pattern or correlation between the "Budget" on the x-axis and "Clicks" on the y-axis
# There's no obvious linear trend - ie, the higher budgets don’t guarantee more clicks - many high-budget campaigns had average or poor conversion
# The data points appear to be randomly scattered across the entire range of both variables, suggesting
# that simply increasing or decreasing the advertising budget does NOT lead to a corresponding increase or decrease in clicks!

plt.figure(figsize=(10, 6))

scatter = sns.scatterplot(
    data= mapp_df,
    x='Budget',
    y='Clicks',
    hue='Conversion_Rate',
    size='Conversion_Rate',
    palette='viridis', # Making it pretty
    alpha=0.7,
    edgecolor='black'
)

plt.title(
    'Advertising Budget vs. Clicks, Colored by Conversion Rate', 
    fontsize=14, 
    weight='bold')

plt.xlabel('Budget')
plt.ylabel('Clicks')

plt.legend(
    title='Conversion Rate', 
    bbox_to_anchor=(1.05, 1), 
    loc='upper left')

plt.tight_layout()
plt.show()
```

**Visualization 4 - Heatmap** - How does ROI vary across different Subscription Tiers and Discount Levels?

```python
# Visualization 4: 

# This heatmap visualizes the average Return on Investment (ROI) for various combinations of Subscription Tiers and Discount Levels
# It helps identify which combinations yield higher or lower profitability...
# For instance, you can quickly spot if a 'Premium' tier with a '10%' discount generally performs better or worse than a 'Basic' tier with the same discount (it does not in particular)
# This can inform pricing strategies and promotional offers!

plt.figure(figsize=(45, 6))

sns.heatmap(roi_pivot, 
            annot=True, 
            fmt=".2f", 
            cmap="coolwarm",  # Making it pretty
            linewidths=0.5,
            annot_kws={"size": 12})

plt.title('Average ROI by Subscription Tier and Discount Level')
plt.xlabel('Discount Level (%)')
plt.ylabel('Subscription Tier')
plt.show()
```

**Visualization 5 - Lineplot** - How does Revenue Generated change over Subscription Length for each Subscription Tier?

```python
# Visualization 5: 

# This line graph illustrates how total revenue generated evolves with increasing subscription length, segmented by different subscription tiers
# It can reveal trends such as whether longer subscriptions consistently bring more revenue
# We can also see if certain tiers see a significant drop or spike in revenue at particular subscription lengths
# This insight is valuable for understanding customer lifetime value and optimizing subscription offerings

plt.figure(figsize=(12, 7))

sns.lineplot(
    data=revenue_by_length_tier,
    x='Subscription_Length',
    y='Revenue_Generated',
    hue='Subscription_Tier',
    marker = 'o',
    palette = 'Set2' # Making it pretty
)

plt.title('Total Revenue Generated by Subscription Length Across Subscription Tiers')
plt.xlabel('Subscription Length (Months)')
plt.ylabel('Total Revenue Generated ($)')
plt.xticks(sorted(mapp_df['Subscription_Length'].unique()))       # Ensures all lengths are shown

plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Subscription Tier')
plt.show()
```

**Visualization 6 - Barplot -** What common keywords are most associated with high-performing products? 

```python
# Visualization 6:

# The graph indicates that 'affordability' is the most common characteristic among high-performing products, followed closely by innovation and durability
# Style also plays some role, but to a slightly lesser extent than the other three
# This data suggests that when aiming for high-performing products, customers value affordability the most, and stylishness the least!

# Convert the Series to DataFrame for seaborn plotting:
keywords_freq_df = most_common_keywords_hp_series.reset_index()
keywords_freq_df.columns = ['Keyword', 'Frequency']

# Plot: 
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Frequency', 
    y='Keyword', 
    data=keywords_freq_df.sort_values(by='Frequency', ascending=True), 
    palette='Spectral', # Making it pretty
    hue='Keyword',  
    legend=False   
)

plt.title(f'Top {num_keywords_to_show} Common Keywords in High-Performing Products')
plt.xlabel('Frequency of Appearance')
plt.ylabel('Keyword')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
